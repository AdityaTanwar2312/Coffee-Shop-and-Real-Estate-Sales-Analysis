```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data\Real_Estate_Sales_2001-2021_GL.csv', low_memory=False)
```

```python
df.info()
```

```python
df.head(5)
```

```python
df['Date Recorded'] = pd.to_datetime(df['Date Recorded'], format = '%m/%d/%Y')
```

```python
df.drop(['Non Use Code', 'Assessor Remarks', 'OPM remarks', 'Location'], axis = 1, inplace = True)
```

```python
df.head(5)
```

```python
df['Residential Type'] = df['Residential Type'].fillna(value = 'Bachelor')
```

```python
df.head(5)
```

```python
df.set_index('Date Recorded')
```

```python
df['Property Type'] = df['Property Type'].fillna(value = 'Other')
```

```python
df['Property Type'].unique()
```

```python
df.reset_index()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Date Recorded', 'Address'], inplace = True)
```
